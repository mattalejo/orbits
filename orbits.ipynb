```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.arange(-2000,2000,0.01)

def rect_orbit(eps, r0, x):
  return np.sqrt((eps**2-1)*x**2 - 2*eps*r0*x + r0**2)

def neg_rect_orbit(eps, r0, x):
  return -np.sqrt((eps**2-1)*x**2 - 2*eps*r0*x + r0**2)


fig,ax = plt.subplots()
ax.set(xlim=(-50, 50), ylim=(-50, 50))
ax.axis('scaled')
ax.plot(0,0)

palette = sns.color_palette(['black'])

color_weight =0.4
color = (color_weight,color_weight,color_weight,1)

for i in range(1,50):
  ax.plot(x,rect_orbit(1,1.8*i,x),linewidth = 1, color = color)
  ax.plot(x,neg_rect_orbit(1,1.8*i,x), linewidth = 1, color = color)
```

```python
fig,ax1 = plt.subplots()
ax1.set(xlim=(-50, 50), ylim=(-50, 50))
ax1.axis('scaled')
ax1.plot(0,0)

for i in range(1,50):
  ax1.plot(x,rect_orbit(0,1.8*i,x),linewidth = 1, color = color)
  ax1.plot(x,neg_rect_orbit(0,1.8*i,x), linewidth = 1, color = color)
```
